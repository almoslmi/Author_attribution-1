```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import os
from sklearn import svm, model_selection, tree, preprocessing, metrics
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import codecs
from nltk.probability import FreqDist
from sklearn import linear_model
from sklearn.model_selection import GridSearchCV
import time
import scipy.special as sp
import random
import helper
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
path='Hindi_train/'
for i in range(5,6):
    try:
        os.mkdir('secondary_features')
    except FileExistsError:
         pass
        
        
    
    f=open('secondary_features/results for '+str(i)+' authors','w')
    iterations=0
    x=int(sp.comb(18,i))
    if(x>20):
        iterations=20
    else:
        iterations=x
        
  
    for a in range(iterations):
         
        random_list=random.sample(range(18),5)
        authors=os.listdir(path);
        temp=[]
        for i in range(len(random_list)):
            temp+=[authors[random_list[i]]]
    
 
        authors=temp

        files=helper.get_files(authors)

        train_data_X,length=helper.add_features(files,None)
        
        le=preprocessing.LabelEncoder()   
        Y=[]
        for author in authors:
            newpath=path+author+'/'
            x=os.listdir(newpath)
            for every_file in x:
                Y=Y+[author]
        train_data_Y=le.fit_transform(Y)
        
        helper.text_normalise(train_data_X,length)  #tf-idf normalisation
        helper.learn(train_data_X,train_data_Y,f)




        

    
    
   # helper.mean_accuracy(iterations,file)
    f.close()


```
